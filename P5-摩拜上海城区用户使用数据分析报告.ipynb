确定订单号没有重复的.并且所有数据项均无null值.主要需要拆分时间和计算租期及直线距离.

```python
#租车时长转换为以秒计量的绝对时间
mobike_original['start_time'] =pd.to_datetime(mobike_original['start_time'])
mobike_original['end_time'] =pd.to_datetime(mobike_original['end_time'])
mobike_original['rent_duration']=(mobike_original['end_time']-mobike_original['start_time']).dt.total_seconds()
mobike_original['rent_duration'].describe()
```

```python
#参考自 https://blog.csdn.net/u013401853/article/details/73368850 ,整理每个订单起始点的直线距离
EARTH_RADIUS=6371           # 地球平均半径，6371km
 
def hav(theta):
    s = sin(theta / 2)
    return s * s
 
def get_distance_hav(lat0, lng0, lat1, lng1):
    "用haversine公式计算球面两点间的距离。"
    # 经纬度转换成弧度
    lat0 = radians(lat0)
    lat1 = radians(lat1)
    lng0 = radians(lng0)
    lng1 = radians(lng1)
 
    dlng = fabs(lng0 - lng1)
    dlat = fabs(lat0 - lat1)
    h = hav(dlat) + cos(lat0) * cos(lat1) * hav(dlng)
    distance = 2 * EARTH_RADIUS * asin(sqrt(h))
 
    return distance
```

```python
#取停车直线距离单位KM
mobike_original['straight_distance']=0.0
for i in mobike_original.index.tolist():
    mobike_original.loc[i,'straight_distance']=get_distance_hav(mobike_original.loc[i,'start_location_x'], mobike_original.loc[i,'start_location_y'], mobike_original.loc[i,'end_location_x'], mobike_original.loc[i,'end_location_y'])
mobike_original['straight_distance'].describe()
```

```python
#pd.__version__
#订单起始时间属于星期0-6,Monday=0, Sunday=6
mobike_original['dayofweek'] = mobike_original['start_time'].dt.weekday_name
mobike_original['dayofweek'].value_counts()
```

```python
#按照上海限行新政,工作日每日7时至10时、15时至20时为高峰期.https://www.guancha.cn/society/2016_04_08_356440.shtml
#https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
mobike_original['start_clock']=mobike_original['start_time'].dt.hour
#mobike_original['start_time'].dt.hour
mobike_original['rush_hour']= 'no' #默认为非高峰时间
for i in mobike_original.index.tolist():
    if mobike_original.loc[i,'start_clock'].hour >=7 and mobike_original.loc[i,'start_clock'].hour < 10:
        mobike_original.loc[i,'rush_hour']= 'morning_rush'
    elif mobike_original.loc[i,'start_clock'].hour >=10 and mobike_original.loc[i,'start_clock'].hour < 15:
        mobike_original.loc[i,'rush_hour']= 'day_time'
    elif mobike_original.loc[i,'start_clock'].hour >=15 and mobike_original.loc[i,'start_clock'].hour < 20:
        mobike_original.loc[i,'rush_hour']= 'afternoon_rush'
    else:
        mobike_original.loc[i,'rush_hour']= 'night_time'
```

```python
mobike_original['start_date']=mobike_original['start_time'].dt.date
mobike_original['start_date'].describe()
```

```python
#单词按半小时(rent_duration=1800s)收费,不满半小时均按半小时计费,因此划分收费范围,ambit_of_charges,1对应最低收费半小时
mobike_original['ambit_of_charges']=mobike_original['rent_duration']//1800 +1
mobike_original['ambit_of_charges'].describe()
```

```python
aggregations = {
    'start_time': {'last_service_date': 'max'},
    'orderid': {'orderid': 'max'}
}
temp_df=mobike_original.groupby('bikeid').agg(aggregations).reset_index()
temp_df.columns = temp_df.columns.droplevel(0)
temp_df.rename({'':'bikeid'},axis=1,inplace=True)
temp_df.set_index(['orderid'], inplace=True)
temp_df.head()
#pd.concat([mobike_original,temp_df],axis=0,join="outer")??
```

```python
mobike_original=pd.concat([mobike_original.set_index(['orderid']),temp_df],axis=1,join_axes=[mobike_original.set_index(['orderid']).index])
```
