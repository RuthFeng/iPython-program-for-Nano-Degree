初步结论:在使用tableau分析之前,需要将原始数据转化为三张可以直接分析的datasheet:按genres分列的一张表,按keywords拆分的一张表,按production_companies拆分的一张表.

```python
#查看数值列是否存在缺失记录或者异常值
movie_df.describe()
```

数值列的count一致,均为10866,说明无空值情况.但是电影时长runtime最小值有为零的,这个按照常识来讲,不合理.因此调查一下这部分数据.

```python
#检查,runtime为0的电影是否存在,记录是否完整
lookuperrer_df=movie_df[movie_df['runtime']== 0]
lookuperrer_df
```

按影片名谷歌搜索,发现对应影片存在,但是其预算和时长等数据严重失真因此清除这些无用条目.

```python
#清理runtime为零的数据列
movie_newdf=movie_df[movie_df['runtime']>0]
movie_newdf.describe()
```

从上述数据看到,剩余影片中runtime最小时长为2分钟,看起来不太合理,拎出来检查下

```python
#检查,runtime为2分钟的电影是否存在,记录是否完整
lookuperrer_df=movie_newdf[movie_newdf['runtime']==2]
lookuperrer_df
```

抽样检查Cousin Ben Troop Screening,发现确实存在,那么runtime的异常清理就到这里了,movie_newdf就为后继分析的基线数据.

```python
#将清理了runtime为零数据后的movie_newdf梳理keywords列分类为'based on novel'和非'based on novel'两类
movie_cleardf=movie_newdf
movie_cleardf.loc[movie_cleardf['keywords'].str.contains("novel",na=False), 'keywords']='based on novel'
movie_cleardf.loc[movie_cleardf['keywords'].str.contains("novel",na=False)==False, 'keywords']='not based on novel'
#将清理了runtime为零数据后的movie_newdf存入movies_cleared.xlsx的'movie_clearbase'页签
writer=pd.ExcelWriter('movies_cleared.xlsx')
movie_cleardf.to_excel(writer,'movie_clearbase', na_rep=0,encoding='utf-8')
```

```python
#定义按关键字对指定列分列处理函数
def splitcolumn(df,colname,keywords):
    try:
        newdf=df
        newcolname='new'+colname
        s=newdf[colname].str.split(keywords).apply(pd.Series,1).stack()
        s.index=s.index.droplevel(-1)
        s.name=newcolname
        del newdf[colname]
        newdf=newdf.join(s)
        return newdf        
    except(ZeroDivisionError,ValueError,TypeError),e:
        print e
```

```python
#梳理genres列,将电影类型分列存储,存入movies_cleared.xlsx的'movie_genressplit'页签
temp_df=splitcolumn(movie_newdf,'genres','|')
temp_df.to_excel(writer,'movie_genressplit', na_rep=0,encoding='utf-8')
```

```python
#梳理production_companies列,将电影厂家分列存储,存入movies_cleared.xlsx的'movie_productionsplit'页签
product_df=splitcolumn(movie_newdf,'production_companies','|')
product_df.to_excel(writer,'movie_productionsplit', na_rep=0,encoding='utf-8')
#保存writer=pd.ExcelWriter('movies_cleared.xlsx')所建文件
writer.save()
```
